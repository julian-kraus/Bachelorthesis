```python
import pandas as pd

df1 = pd.read_csv("interpreted_data_mlx_J.csv")
df2 = pd.read_csv("interpreted_data_mlx_part2_j.csv")
df3 = pd.read_csv("interpreted_data_mlx_test2.csv")
df4 = pd.read_csv("interpreted_data_mlx_test2_2.csv")
df1.head()
```

```python
df2.head
```

```python
df1 = df1.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')
df2 = df2.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')
df3 = df3.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')
df4 = df4.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')
```

```python
df1.iloc[-len(df2):] = df2.values
df3.iloc[-len(df4):] = df4.values
```

```python
dfs = [df1, df3]
df = pd.concat(dfs)
print(len(df))
```

```python
meniskus = [
    "M23.3",
    "S83.2", # Rupture
]
meniskus_low = [
    "M23.0", # Ganglion doesn't seem to urgent
    "M23.1", # something you have from birth
    "M23.2", # problems because of old rupture
    "M23.9"  # Same as M23.8
]
cruciate = [
    "M23.6", # Ruptur
    "S83.50", # might be rupture
    "S83.53", # Rupture
    "S83.54", # Rupture
    "S83.7" # Multiple areas
]
cruciate_low = [
    "M23.8", # Seems to be no rupture, rather weakness, still operations
    "S83.51", # Distorsion
    "S83.52", # Distorsion
]
other_urgent = [
    "S83.3" # Rupture Kniegelenkknorpel
    "S83.4" # Rupture Seitenband
]
other = [
    "M17",
    "M22",  
    "M23.4", # Freikörper, does have a lot of operations
    "M23.5", # instability
    "S83.1", # Only luxation
    "S83.6" # Distorsion somewhere
]
def createLabel(code):
    if any(elem in code for elem in meniskus):
        return "meniskus-urgent"
    elif any(elem in code for elem in meniskus_low):
        return "meniskus-minor"
    elif any(elem in code for elem in cruciate):
        return "cruciate-ligament-urgent"
    elif any(elem in code for elem in cruciate_low):
        return "cruciate-ligament-minor"
    elif any(elem in code for elem in other_urgent):
        return "other-urgent"
    elif any(elem in code for elem in other):
        return "other-minor"
    else:
        return "other-minor"
    
df['label'] = df['DIA_code'].apply(createLabel)
```

```python
import numpy as np
import pandas as pd

# Function to translate sex
def translate_sex(sex):
    if sex == 'male':
        return 'männlich '
    elif sex == 'female':
        return 'weiblich '
    else:
        return ""
    
def format_age(age):
    if not pd.isna(age):
        return str(age) + " Jahre alt "
    else:
        return ""

cols = ['ANA', 'EXA']

# Ensure age and sex are included in the DataFrame and handle NaN values
df['data'] = df.apply(lambda row: (format_age(row['age']) + translate_sex(row['sex']) + ". ".join(row[cols].dropna().astype(str))
), axis=1)
df['data2'] = df.apply(lambda row: ". ".join(row[cols].astype(str)), axis=1)
```

```python
df = df.dropna(subset=['ANA', 'EXA'], how='all')   
df = df[df['DIA_code'].notna()]
len(df)
```

```python
df.to_csv("full_interpreted.csv")
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(df, test_size=0.4, random_state=42, stratify=df['label'])

# Split the temporary data into validation and test sets (each 50% of temporary, thus 20% of total each)
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['label'])
```
