```python
import IO
import pandas as pd
```

```python
type = "Meniskus"
bef = IO.load("./data/" + type + "_BEF.xls")
an = IO.load("./data/" + type + "_ANA.xls")
print(an.shape[0])
print(bef.shape[0]) 
an.head()
```

```python

ids_df1 = set(bef['Patientid'])
ids_df2 = set(an['Patientid'])
if ids_df1 == ids_df2:
    print("Both dataframes have the same IDs.")
else:
    print("The IDs do not match.")
    missing_in_df1 = ids_df2 - ids_df1
    missing_in_df2 = ids_df1 - ids_df2
    print(f"IDs in df2 not in df1: {missing_in_df1}")
    print(f"IDs in df1 not in df2: {missing_in_df2}")
print(an.shape[0])
print(bef.shape[0]) 
an.head()
```

```python
men = pd.merge(bef, an, left_index=True, right_index=True)
print(men.shape[0])
```

```python
men["Diagnose"] = type
men.head()
```

```python
men.set_index("Patientid_x", inplace=True)
men.rename(columns={"Text_x": "Text_Bef", "Text_y": "Text_ANA", "Geschlecht_x": "Geschlecht"}, inplace=True)
men = men[['Text_Bef', 'Text_ANA', 'Geschlecht', 'Diagnose']]
men.dropna()
```

```python
men.to_excel("../data/" + type + ".xlsx", engine='xlsxwriter')
```
